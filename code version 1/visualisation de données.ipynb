### Rayan: Résultats liées au Basketball (résultats en fonction des jours)

```python
# Visualiser les performances de Rayan en fonction des jours avec un graphique
# Rayan est positionné à 5 pas d'un panier de basketball. 
# Il doit effectuer 10 tirs, d'où vient sa performance (la quantité de bons tirs sur les 10 tirs)

sns.set(font_scale=1.2, font="Verdana")

plt.figure(figsize=(10, 6))
sns.lineplot(x=dates, y=athlete_results_rayan)

plt.xlabel("Dates de l'activité physique", fontweight="bold", fontsize=10)
plt.ylabel("Performance de Rayan (taux de bons tirs)", fontweight="bold", fontsize=10)

plt.title("Variation des performances de Rayan en fonction des jours", fontweight="bold", fontsize=14)

plt.xticks(rotation=45)
plt.show()
```

```python
# Visualiser les valeurs de Alpha α en fonction des jours avec un graphique
plt.figure(figsize=(10, 6))
sns.lineplot(x=dates, y=alpha_values)

plt.xlabel("Dates de l'activité physique", fontweight="bold")
plt.ylabel("Valeurs de Alpha α", fontweight="bold")

plt.title("Variation des valeurs de Alpha α en fonction des jours", fontweight="bold", fontsize=14)

plt.xticks(rotation=45)
plt.show()
```

```python
# Graphique Commun 
sns.set(font_scale=1.05, font="Verdana")

plt.figure(figsize=(10, 6))

# Plotting Rayan's performance
rayan_line = sns.lineplot(
    x=dates, 
    y=athlete_results_rayan, 
    label="Performances de Rayan",
    style=True,
    dashes=[(2, 2)],
    legend=False
)

# Plotting Alpha values
alpha_plot = sns.lineplot(
    x=dates, 
    y=alpha_values, 
    label="Valeurs de Alpha α",
    style=True,
    dashes=[(2, 2)],
    legend=False
)

plt.xlabel("Dates de l'activité physique", fontweight="bold")
plt.ylabel("Performances de Rayan", fontweight="bold")

plt.title("Variation des performances de Rayan et des valeurs de Alpha α en fonction des jours", fontweight="bold", fontsize=14)

# Adding titles to each point on the Alpha graph
for i in range(len(dates)):
    alpha_plot.text(dates[i], alpha_values[i], particle_size[i], horizontalalignment='left', size='medium', color='black')

plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Yara: Résultats liées aux Jumping Jacks (résultats en fonction des jours) 

```python
sns.set(font_scale=1.2, font="Verdana")

plt.figure(figsize=(16, 32))
for i in range(8):
    plt.subplot(8, 1, i + 1)
    sns.lineplot(x=x_labels, y=[athlete_results_yara_restingBPM[i],
                                athlete_results_yara_after15s_exercise[i],
                                athlete_results_yara_after15s_rest[i],
                                athlete_results_yara_after30s_exercise[i],
                                athlete_results_yara_after30s_rest[i],
                                athlete_results_yara_after60s_exercise[i],
                                athlete_results_yara_after60s_rest[i]])
    plt.title(f"{dates[i]} (α = {alpha_values[i]}, {particle_size[i]})", fontweight="bold", fontsize=23)
    plt.ylabel("BPM de Yara", fontweight="bold", fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Graphique Commun 
sns.set_theme(font_scale=1.12, font="Verdana")
fig, ax1 = plt.subplots(figsize=(12, 8))

# Créer le second axe
ax2 = ax1.twinx()

# Graphique des Performances de Yara (pour le premier graphique des y)
yara_line = sns.lineplot(
    x=dates, y=differences_yara,
    ax=ax1, style=True, 
    color='b', dashes=[(2, 2)],
    label="Performance de Yara (BPM)", legend=False
)
yara_line.legend(loc="upper right")

# Graphique des valeurs de Alpha (pour le deuxième graphique des y)
alpha_line = sns.lineplot(
    x=dates, y=alpha_values,
    ax=ax2, style=True, 
    color='r', dashes=[(2, 2)],
    label="Valeurs de Alpha α", legend=False
)
alpha_line.legend(loc="upper left")

# Tailles des particules
for i in range(len(dates)):
    ax2.text(dates[i], alpha_values[i], particle_size[i], horizontalalignment='left', size='small', color='black')

# Les labels
ax1.set_xlabel("Dates de l'activité physique", fontweight="bold")
ax1.set_ylabel("Performance de Yara (BPM)", fontweight="bold")
ax2.set_ylabel("Valeurs de Alpha α", fontweight="bold")

# Le titre du graphique
plt.title("Variation des performances de Yara & des valeurs de Alpha α en fonction des jours", fontweight="bold", fontsize=14)

# Rotation de l'axe des x
plt.xticks(rotation=45)

# Combiner les légendes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

plt.show()
```
