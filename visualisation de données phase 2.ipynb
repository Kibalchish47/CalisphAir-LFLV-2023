# Analyse des données 

#### Matrice de Covariance
La covariance est une mesure qui indique le degré auquel deux variables varient ensemble. Si nous avons un ensemble de **n** variables, alors la matrice de covariance est une matrice **n x n** où l'élément à l'intersection de la **i-ème** ligne et de la **j-ème** colonne est la covariance entre la **i-ème** et la **j-ème** variable de l'ensemble de données. 

La formule pour calculer la covariance entre deux ensembles de valeurs **X** et **Y** est :

$$ Cov(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n-1} $$

où $ n $ est le nombre d'éléments dans $ X $ et $ Y $, $ X_i $ et $ Y_i $ sont les éléments individuels de $ X $ et $ Y $, et $\bar{X}$ et $\bar{Y}$ sont les moyennes de $ X $ et $ Y $ respectivement.

#### Corrélation de Pearson
C'est une mesure de la relation linéaire entre deux variables aléatoires. Elle varie entre -1 et 1. Une valeur proche de 1 indique une forte corrélation positive, une valeur proche de -1 indique une forte corrélation négative, et une valeur proche de 0 indique une absence de corrélation. 

La formule pour calculer la corrélation de Pearson est :

$$ r_{xy} = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n} (X_i - \bar{X})^2 \sum_{i=1}^{n} (Y_i - \bar{Y})^2}} $$

#### Corrélation de Spearman
C'est une mesure de la relation monotonique entre deux variables. Elle est basée sur les rangs des valeurs plutôt que sur les valeurs elles-mêmes. Comme la corrélation de Pearson, elle varie entre -1 et 1. 

La formule pour calculer la corrélation de Spearman est :

$$ r_s = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} $$

où $ d_i $ est la différence entre les rangs de chaque observation, et $ n $ est le nombre total d'observations.

```python
# Graphique Commun 
sns.set_theme(font_scale=1.12, font="Verdana")
fig, ax1 = plt.subplots(figsize=(12, 8))

# Créer le second axe
ax2 = ax1.twinx()

# Graphique des Performances de Yara (pour le premier graphique des y)
yara_line = sns.lineplot(
    x=dates, y=differences_yara,
    ax=ax1, style=True, 
    color='b', dashes=[(2, 2)],
    label="Performance de Yara (BPM)", legend=False
)
yara_line.legend(loc="upper right")

# Graphique des valeurs de Alpha (pour le deuxième graphique des y)
alpha_line = sns.lineplot(
    x=dates, y=alpha_values,
    ax=ax2, style=True, 
    color='r', dashes=[(2, 2)],
    label="Valeurs de Alpha α", legend=False
)
alpha_line.legend(loc="upper left")

# Tailles des particules
for i in range(len(dates)):
    ax2.text(dates[i], alpha_values[i], particle_size[i], horizontalalignment='left', size='small', color='black')

# Les labels
ax1.set_xlabel("Dates de l'activité physique", fontweight="bold")
ax1.set_ylabel("Performance de Yara (BPM)", fontweight="bold")
ax2.set_ylabel("Valeurs de Alpha α", fontweight="bold")

# Le titre du graphique
plt.title("Variation des performances de Yara & des valeurs de Alpha α en fonction des jours", fontweight="bold", fontsize=14)

# Rotation de l'axe des x
plt.xticks(rotation=45)

# Combiner les légendes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

plt.show()

print("Différences entre le BPM juste après 60s d'Activité et juste après 60s de Repos de Yara.")
print("Différences = ", differences_yara)
```

### Correlation: Performance Yara / Alpha α

```python
print("YARA")
correlation_datasets = {"Différences BPM après 60s d'Activité et après 60s de Repos (Yara)": differences_yara, 
                        "BPM Repos (Yara)": athlete_results_yara_restingBPM, 
                        "BPM Après 20s d'Activité (Yara)": athlete_results_yara_after20s_exercise, 
                        "BPM Après 20s de Repos (Yara)": athlete_results_yara_after20s_rest, 
                        "BPM Après 40s d'Activité (Yara)": athlete_results_yara_after40s_exercise, 
                        "BPM Après 40s de Repos (Yara)": athlete_results_yara_after40s_rest,
                        "BPM Après 60s d'Activité (Yara)": athlete_results_yara_after60s_exercise, 
                        "BPM Après 60s de Repos (Yara)": athlete_results_yara_after60s_rest}

for name, dataset in correlation_datasets.items():
    print(f"\nCorrelation entre Alpha α et {name}: \n")
    
    # Calculer la matrice de covariance
    covariance_yara = np.cov(dataset, alpha_values)
    print("Matrice de Covariance: \n", covariance_yara)
    
    # Calculer et quantifier la correlation de Pearson
    corr_pearson_yara, _ = scipy.stats.pearsonr(dataset, alpha_values)
    print("Correlation de Pearson: %.3f" % corr_pearson_yara)
    
    # Calculer et quantifier la correlation de Spearman
    corr_spearman_yara, _ = scipy.stats.spearmanr(dataset, alpha_values)
    print("Correlation de Spearman: %.3f" % corr_spearman_yara)
```

### Correlation: Performance Aya / Alpha α

```python
print("AYA")
correlation_datasets = {"Différences BPM après l'Activité et après le Repos (Aya)": differences_aya, 
                        "BPM Repos (Aya)": athlete_results_aya_restingBPM, 
                        "BPM Après Activité (Aya)": athlete_results_aya_after_exercise, 
                        "BPM Après Repos (Aya)": athlete_results_aya_after_rest}

for name, dataset in correlation_datasets.items():
    print(f"\nCorrelation entre Alpha α et {name}: \n")
    
    # Calculer la matrice de covariance
    covariance_aya = np.cov(dataset, alpha_values)
    print("Matrice de Covariance: \n", covariance_aya)
    
    # Calculer et quantifier la correlation de Pearson
    corr_pearson_aya, _ = scipy.stats.pearsonr(dataset, alpha_values)
    print("Correlation de Pearson: %.3f" % corr_pearson_aya)
    
    # Calculer et quantifier la correlation de Spearman
    corr_spearman_aya, _ = scipy.stats.spearmanr(dataset, alpha_values)
    print("Correlation de Spearman: %.3f" % corr_spearman_aya)
```
